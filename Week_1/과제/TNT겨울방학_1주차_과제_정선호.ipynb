```
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/과제/climate_change.csv')


print('\n')
print(df.info())
```

**1-2**

```

fig = plt.figure(figsize=(10,10))
date = df['date']
co2= df['co2']
temp = df['relative_temp']

ax1 = fig.add_subplot(2,1,1)
ax2 = fig.add_subplot(2,1,2)
ax1.set_xlabel('Time')
ax1.set_ylabel('CO2')
ax1.plot(date,co2, color='blue')

ax2.set_xlabel('Time')
ax2.set_ylabel('Temp')
ax2.plot(date,temp, color='red')



ax1.set_xticklabels(date,rotation=90)
ax2.set_xticklabels(date, rotation=90)
plt.show()
```
